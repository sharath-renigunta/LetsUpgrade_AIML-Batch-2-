```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data_f = pd.read_csv('mtcars.csv')
data_f
```

```python
data_f.describe()
```

```python
data = list(data_f['hp'])
```

```python
data
```

```python
print(sns.boxplot(data))
```

```python
mean=np.mean(data)
print(mean)
sd=np.std(data)
print(sd)
print(data)
```

```python
def outlier(data):
    thresholdValue = 2
    outlier = []
    for i in data:
        z_test = (i - mean) / sd
        if z_test > thresholdValue:
            outlier.append(i)
    print(outlier)
```

```python
outlier(data)
```
